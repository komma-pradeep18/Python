```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

color = pd.DataFrame({
    'Color Name': ['Red', 'Green', 'Blue', 'Yellow', 'Black', 'White'],
    'Hex Code': ['#FF0000', '#008000', '#0000FF', '#FFFF00', '#000000', '#FFFFFF']
})

print(color)

label_encoder = LabelEncoder()
color['Color_label'] = label_encoder.fit_transform(color['Color Name'])

print(color.head())

one_hot_encoder = OneHotEncoder()
one_hot_encoded = one_hot_encoder.fit_transform(color[['Color Name']])
one_hot_encoded_df = pd.DataFrame(one_hot_encoded.toarray(), columns=one_hot_encoder.get_feature_names_out())

color = pd.concat([color, one_hot_encoded_df], axis=1)
print(color)
```
